```python
# prompt: buatkan line chart untuk semua jurusan

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("/content/tuition_graduate.csv")

# Get unique school names
schools = data['school'].unique()

# Create a line chart for each school
plt.figure(figsize=(15, 10))

for school in schools:
    school_data = data[data['school'] == school]
    plt.plot(school_data['academic.year'], school_data['cost'], marker='o', linestyle='-', label=school)

plt.xlabel('Academic Year')
plt.ylabel('Cost')
plt.title('Cost Over Time for All Schools')
plt.grid(True)
plt.legend()  # Show legend for different schools
plt.xticks(rotation=45, ha="right") # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# prompt: buatkan pie chart untuk feature "Summary"

import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'Summary' is a column in it
summary_counts = data['school'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(summary_counts, labels=summary_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Summaries')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
