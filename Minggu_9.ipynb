# **Minggu 9: Visualisasi Data Lanjutan**

Boxplot: Menunjukkan siapa yang nilainya tinggi, rata-rata, atau rendah

Heatmap: Melihat pola nilai siswa secara keseluruhan

Pairplot: Apakah ada hubungan antar mata pelajaran?

Histogram: Apakah nilai siswa banyak di atas rata-rata?

```python
# âœ¨ Ayo Mulai! Kita akan belajar membuat grafik keren dari data sederhana.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
%matplotlib inline
```

##**1. Data Nilai Siswa**

```python
# Kita buat data sederhana: nilai 10 siswa dalam 3 mata pelajaran

data = {
    "Nama": ["Ani", "Budi", "Citra", "Deni", "Eka", "Fajar", "Gita", "Hana", "Ivan", "Joko"],
    "Matematika": [85, 78, 92, 60, 70, 88, 75, 80, 90, 65],
    "Fisika": [80, 70, 95, 65, 75, 85, 78, 82, 88, 60],
    "Kimia": [78, 72, 88, 58, 76, 90, 74, 79, 86, 62]
}

df = pd.DataFrame(data)
df
```

##**ðŸ“ˆ 2. Boxplot: Perbandingan Nilai**

```python
# Kita lihat persebaran nilai tiap mata pelajaran
df_melted = df.melt(id_vars="Nama", var_name="Mata Pelajaran", value_name="Nilai")

sns.boxplot(x="Mata Pelajaran", y="Nilai", data=df_melted, palette="Set2")
plt.title("Persebaran Nilai Siswa per Mata Pelajaran")
plt.show()
```

##**ðŸ”¥ 3. Heatmap: Siapa Nilainya Mirip?**

```python
# Kita lihat siapa siswa yang punya pola nilai mirip
nilai_siswa = df.drop(columns="Nama")
sns.heatmap(nilai_siswa, annot=True, cmap="YlGnBu")
plt.title("Heatmap Nilai Siswa")
plt.xlabel("Mata Pelajaran")
plt.ylabel("Siswa")
plt.show()
```

##**ðŸ” 4. Pairplot: Apakah yang jago Matematika juga jago Kimia?**

```python
sns.pairplot(nilai_siswa)
plt.suptitle("Hubungan antar Nilai Mata Pelajaran", y=1.02)
plt.show()
```

##**ðŸ“Š 5. Histogram: Distribusi Nilai**

```python
sns.histplot(df["Matematika"], kde=True, color="green")
plt.title("Distribusi Nilai Matematika")
plt.xlabel("Nilai")
plt.ylabel("Jumlah Siswa")
plt.show()
```

```python
# 4. Distribusi Data: Histogram dan KDE
print("Distribusi Data - Histogram & KDE")
sns.histplot(df["petal_length"], kde=True, color="purple")
plt.title("Distribusi Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# 5. Grafik Multi-Panel (Subplot)
print("Grafik Multi-Panel")
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.histplot(df["sepal_length"], kde=True, ax=axes[0,0], color="skyblue")
axes[0,0].set_title("Distribusi Sepal Length")

sns.boxplot(x="species", y="sepal_width", data=df, ax=axes[0,1], palette="pastel")
axes[0,1].set_title("Boxplot Sepal Width")

sns.violinplot(x="species", y="petal_length", data=df, ax=axes[1,0], palette="muted")
axes[1,0].set_title("Violin Plot Petal Length")

sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df, ax=axes[1,1])
axes[1,1].set_title("Scatter Plot Sepal vs Petal Length")

plt.tight_layout()
plt.show()
```
